# Warshall algorithm to find reachability between all pairs of vertices in a graph.
Problem Description:
Understand transitive closure of a graph. Use Python to compute reachability using matrix method.

#Theory
Warshall algorithm finds whether a path exists between every pair of vertices. It updates the adjacency matrix step by step. If there is a path from i to j through k, then the entry becomes 1.

## Algorithm
1. Represent graph using adjacency matrix.
2. For each intermediate vertex k
3. Check all pairs i and j
4. Update value if path exists through k
4. Repeat for all vertices


```python
def warshall(graph):

    n = len(graph)

    for k in range(n):
        for i in range(n):
            for j in range(n):
                graph[i][j] = graph[i][j] or (graph[i][k] and graph[k][j])

    return graph
```

```python
def warshall(graph):
    n = len(graph)
    reach = [row[:] for row in graph]

    for k in range(n):
        for i in range(n):
            for j in range(n):
                reach[i][j] = reach[i][j] or (reach[i][k] and reach[k][j])
    return reach


def floyd(graph):
    n = len(graph)
    dist = [[float('inf')] * n for _ in range(n)]


    for i in range(n):
        for j in range(n):
            if graph[i][j] != 0:
                dist[i][j] = graph[i][j]
        dist[i][i] = 0


    for k in range(n):
        for i in range(n):
            for j in range(n):
                if dist[i][k] + dist[k][j] < dist[i][j]:
                    dist[i][j] = dist[i][k] + dist[k][j]
    return dist



n = 5
print("Enter adjacency matrix (5x5):")
graph = []
for i in range(n):
    row = list(map(int, input().split()))
    graph.append(row)


closure = warshall(graph)
print("\nTransitive Closure Matrix (Warshall):")
for row in closure:
    print(row)


shortest_paths = floyd(graph)
print("\nShortest Path Matrix (Floyd):")
for row in shortest_paths:
    print(row)


print("\nPath Existence (Warshall):")
for i in range(n):
    for j in range(n):
        if closure[i][j]:
            print(f"Path exists from {i+1} → {j+1}")
        else:
            print(f"No path from {i+1} → {j+1}")
```

## Find shortest paths between all pairs of vertices using Floyd Warshall algorithm.

#Understand all pairs shortest path problem. Implement dynamic programming approach in Python.

* Given a weighted graph with positive or negative edge weights, find
shortest distances between every pair of vertices.

# Theory
Floyd Warshall algorithm computes shortest paths between all pairs of nodes. It updates the distance matrix by checking if a shorter path exists through an intermediate vertex. Works with negative weights but fails with negative cycles.

Algorithm
1. Represent graph using distance matrix.
2. Initialize matrix with given weights.
3. Set distance of node to itself as 0.
4. For each intermediate vertex k
5. Update distance[i][j] = min(distance[i][j], distance[i][k] + distance[k][j])
6. Repeat for all vertices

```python
def floyd_warshall(graph):

    n = len(graph)

    dist = []
    for i in range(n):
        dist.append(graph[i][:])

    for k in range(n):
        for i in range(n):
            for j in range(n):
                if dist[i][k] + dist[k][j] < dist[i][j]:
                    dist[i][j] = dist[i][k] + dist[k][j]

    return dist

```

```python
INF = float('inf')

graph = [
[0, 3, INF, 7],
[8, 0, 2, INF],
[5, INF, 0, 1],
[2, INF, INF, 0]
]

result = floyd_warshall(graph)

print("Shortest distance matrix:")

for row in result:
    print(row)
```

```python
import heapq

INF = float('inf')

def floyd_warshall(graph):
    n = len(graph)
    dist = [row[:] for row in graph]

    pred = [[None if graph[i][j] == INF else i for j in range(n)] for i in range(n)]

    for k in range(n):
        for i in range(n):
            for j in range(n):
                if dist[i][k] != INF and dist[k][j] != INF:
                    if dist[i][k] + dist[k][j] < dist[i][j]:
                        dist[i][j] = dist[i][k] + dist[k][j]
                        pred[i][j] = pred[k][j]

    for i in range(n):
        if dist[i][i] < 0:
            print("Graph contains a negative cycle!")
            return dist, pred, True

    return dist, pred, False

def reconstruct_path(pred, i, j):
    if pred[i][j] is None:
        return []
    path = [j]
    while j != i:
        j = pred[i][j]
        path.append(j)
    path.reverse()
    return path

def dijkstra(graph, src):
    n = len(graph)
    dist = [INF] * n
    dist[src] = 0
    pq = [(0, src)]
    while pq:
        d, u = heapq.heappop(pq)
        if d > dist[u]:
            continue
        for v in range(n):
            if graph[u][v] != INF:
                if dist[u] + graph[u][v] < dist[v]:
                    dist[v] = dist[u] + graph[u][v]
                    heapq.heappush(pq, (dist[v], v))
    return dist

n = int(input("Enter number of vertices: "))
print(f"Enter weighted adjacency matrix for {n} vertices (use {INF} for no edge):")
graph = []
for i in range(n):
    row = list(map(float, input().split()))
    if len(row) != n:
        raise ValueError(f"Row {i+1} must have {n} entries")
    graph.append(row)

dist_fw, pred_fw, neg_cycle = floyd_warshall(graph)
print("\nFloyd-Warshall Shortest Distance Matrix:")
for row in dist_fw:
    print(row)

if not neg_cycle:
    u, v = map(int, input("\nEnter source and destination vertices to find shortest path: ").split())
    path = reconstruct_path(pred_fw, u, v)
    if path:
        print(f"Shortest path from {u} to {v}: {path} with distance {dist_fw[u][v]}")
    else:
        print(f"No path exists from {u} to {v}")

src = int(input("\nEnter source vertex for Dijkstra: "))
dist_dij = dijkstra(graph, src)
print(f"Dijkstra shortest distances from {src}: {dist_dij}")

```
